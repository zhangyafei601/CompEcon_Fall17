# Problem Set 3

# Step 1: Select data

```python
# import packages we'll use here
import pandas as pd
import numpy as np
import scipy.optimize as opt
import time
```

```python
# Read in the data
ps3_data = pd.read_stata('PS3_data.dta')
```

```python
ps3_data.head(n=5)
```

```python
ps3_data.describe()
```

```python
# Drop missing observations for income and hours and observations with hours =0
ps3_data = ps3_data[(ps3_data.hlabinc.isnull() != True ) & (ps3_data.hannhrs.isnull() != True ) & (ps3_data.hannhrs > 0)]
```

```python
ps3_data = ps3_data[(ps3_data.hyrsed.isnull() != True) & (ps3_data.age.isnull() != True) & (ps3_data.hrace.isnull() != True)]
```

```python
# Create log value of wage per hour
ps3_data['wage_h'] = ps3_data['hlabinc'] / ps3_data['hannhrs'] 
ps3_data['wage_h_log'] = np.log(ps3_data['wage_h'])
```

```python
ps3_data.describe()
```

```python
# Select only male heads of household who are between 25 and 60 years old and earn wages > $7/hr
ps3_data = ps3_data[(ps3_data.hsex ==1 ) & (ps3_data.age >= 25) & (ps3_data.age <= 60) & (ps3_data.wage_h > 7)]
```

```python
ps3_data.describe()
```

```python
ps3_data.head(n=5)
```

# Step 2: Create variables

```python
# Create three dummy variables. 1 means true, 0 means false
ps3_data['Black'] = ps3_data.loc[:,'hrace'].apply(lambda x : 1 if x == 2 else 0)
ps3_data['Hispanic'] = ps3_data.loc[:,'hrace'].apply(lambda x : 1 if x == 5 else 0)
ps3_data['OtherRace'] = ps3_data.loc[:,'hrace'].apply(lambda x : 1 if x == 3 or x == 4 or x == 6 or x == 7 else 0)

# Create constant variable
ps3_data['constant'] = 1
```

```python
ps3_data.head(n=5)
```

# Step 3: Estimate MLE

## Estimate the model for 1971

## (1) Select data of 1971

```python
ps3_data_1971 = ps3_data[(ps3_data['year'] == 1971)]
```

```python
ps3_data_1971.describe()
```

## (2) Define log-likelihood function

```python
# Assuming normal distribution of the error term, the log-likelihood function below is given by the book of Hayashi
def loglike(params, x, y):
    
    beta = np.array([[params[0]], [params[1]], [params[2]], [params[3]], [params[4]]], dtype = np.float64)
    
    x = np.array([ps3_data_1971['constant'], ps3_data_1971['hyrsed'], ps3_data_1971['age'], 
                  ps3_data_1971['Black'], ps3_data_1971['OtherRace']]).T
    
    y = np.array([ps3_data_1971['wage_h_log']]).T
      
    sigma = 2 # The value of sigma is not sensitive, different values of sigma will yield exactly the same results
    
    params = [alpha, beta1, beta2, beta3, beta5]
    
    log_like = (- (-(1 / 2) * len(ps3_data_1971.index) * np.log(2 * (np.pi)) - len(ps3_data_1971.index) * np.log(sigma) - 
                  (1 / 2) * np.dot((y - np.dot(x, beta)).T, y - np.dot(x, beta)) / ((sigma) ** 2))).item((0,0))
    
    return log_like
```

## (3) Call the minimizer

```python
# Initial guesses at the parameter values
alpha, beta1, beta2, beta3, beta5 = 0.1, -0.1, 0.2, -0.1, 0.5
params = [alpha, beta1, beta2, beta3, beta5]
params_initial = (0.3, -0.5, 0.6, -0.3, 0.8)

# Call minimzer to minimize the log-likelihood function
mle_results = opt.minimize(loglike, params_initial, args=(ps3_data_1971,params),
                           method='Nelder-Mead', tol = 1e-12, options={'maxiter': 5000})

mle_results
```
