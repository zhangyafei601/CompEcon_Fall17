```python
# Assuming normal distribution of the error term, the log-likelihood function below is given by the book of Hayashi
def loglike(params, year, dataset):
    
    
    z=np.array([list(params)])
    beta = np.transpose(z)
    
    #n returns the sum of obs that have a value equal to the parameter year
    n = np.sum(dataset.year == year)
    
    #Note [set3['cons'][set3.year==year] >>>> This extract a portion of the data set
    # that has the column 'year' = to the parameter year, i.e. a portion of the matrix. 
    #The same for other columns...
    
    
    x = np.array([dataset['cons'][dataset.year == year], dataset['hyrsed'][dataset.year == year],
                  dataset['age'][dataset.year == year], dataset['Black'][dataset.year == year], 
                  dataset['OtherRace'][dataset.year == year]]).T
    
    y = np.array([dataset['logwage'][dataset.year == year]]).T
    
    epsilon = y - np.dot(x, beta)
    
    essilon_s = np.dot(epsilon, epsilon.T)
    
    sigma = test(dataset, year, params)
    
    
    
    
   
    
    log_like = (((n/2) * np.log(2 * (np.pi))) + ((n)/2)* np.log((sigma**2)) +
                  (( essilon_s / ((sigma ** 2))/2))).item(0,0)
    
    return log_like
```

```python
def test(dataset, year, params):
    
    z=np.array([list(params)])
    beta = np.transpose(z)
    n = np.sum(dataset.year == year)
    x = np.array([dataset['cons'][dataset.year == year], dataset['hyrsed'][dataset.year == year],
                  dataset['age'][dataset.year == year], dataset['Black'][dataset.year == year], 
                  dataset['OtherRace'][dataset.year == year]]).T
    
    y = np.array([dataset['logwage'][dataset.year == year]]).T
    
    var = np.dot((y - np.dot(x, beta)).T, (y - np.dot(x, beta)) ) / n
                 

    return var.item(0,0)
```

```python
params0 = (0.1, -0.1, 0.2, -0.1, 0.6)
bnds = ((None, None), (None, None), (None, None),
        (None, None), (None, None))
year=1971
dataset=set3
# Call minimzer to minimize the log-likelihood function
mle_results = opt.minimize(loglike, params0,args=(year, dataset),
                           method='SLSQP',bounds=bnds, tol=1e-15,)

mle_results
```
